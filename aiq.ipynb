```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
encoded_cities = le.fit_transform(data['City'])
```

```python
import pickle
city_mapping = dict(zip(data['City'], encoded_cities))
with open('city_mapping.pkl', 'wb') as file:
    pickle.dump(city_mapping, file)

print("City Mapping Saved:", city_mapping)
```
